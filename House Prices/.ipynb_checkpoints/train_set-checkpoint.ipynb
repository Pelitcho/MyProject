### D. Missing values

```python
def count_miss(df):
    l_miss=[]
    for i in df.columns:
        if df[i].isnull().sum()>0:
            l_miss.append(i)
    return l_miss
```

```python
l_miss=count_miss(df)
```

```python
from sklearn import linear_model
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.cross_decomposition import PLSRegression
from sklearn.ensemble import GradientBoostingRegressor
from sklearn import tree
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.neural_network import MLPRegressor

KFold = KFold(n_splits=10, shuffle=True, random_state=42)

models = []
models.append(('LR', linear_model.LinearRegression()))
models.append(('LNR', linear_model.Ridge()))
models.append(('BAY', linear_model.BayesianRidge()))
models.append(('RANSAC', linear_model.RANSACRegressor()))
models.append(('SDGR', linear_model.SGDRegressor()))
models.append(('KN', KNeighborsRegressor()))
models.append(('DTR', tree.DecisionTreeRegressor()))
models.append(('SVM', SVR()))
models.append(('PLSR', PLSRegression()))
models.append(('GBR', GradientBoostingRegressor()))
models.append(('NN', MLPRegressor()))

results_models=[]

names=[]

for name,model in models:
    cv_results= cross_val_score(model, X_train, y_train.values.ravel(),cv=KFold,scoring='r2')
    results_models.append(cv_results)
    names.append(name)
    print('%s: %f (%f)' %(name,cv_results.mean(),cv_results.std()))
```

### B.Best_model

```python
plt.boxplot(results_models,labels=names)
plt.title('Algorithm Comparison')
plt.ylim(0.20,0.90)
plt.xticks(rotation=45)
plt.show()
```

The best model here is SVM wich got the minimum value of std.

### C.Predictions

```python
model_final=SVR()
model_final.fit(X_train,y_train.values.ravel())
y_pred=model_final.predict(X_test)
```

```python
#Evaluation predictions
from sklearn import metrics
print(metrics.r2_score(y_test, y_pred))
print(metrics.mean_squared_error(y_test,y_pred))
print(metrics.mean_absolute_error(y_test,y_pred))
```

### D.Optimisation

```python
from time import time
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001]}

grid = GridSearchCV(model_final, param_grid, refit = True, scoring='r2',cv=KFold,verbose=3)
grid.fit(X_train,y_train.values.ravel())
start=time()
print(
    "GridSearchCV took %.2f seconds for %d candidate parameter settings."
    % (time() - start, len(grid.cv_results_["params"]))
)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results["rank_test_score"] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print(
                "Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results["mean_test_score"][candidate],
                    results["std_test_score"][candidate],
                )
            )
            print("Parameters: {0}".format(results["params"][candidate]))
            print("")
report(grid.cv_results_)
```
