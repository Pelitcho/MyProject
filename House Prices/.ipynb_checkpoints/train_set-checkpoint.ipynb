```python
import numpy as np
def sigmoid(z):
    return 1/(1+np.exp(-z))
```

```python
def relu(x):
	return np.maximum(0.0, x)
```

```python
def loss(y,a):
    return  -(y*np.log(a)+(1-y)*np.log(1-a))
```

```python
def mse(y,a):
    return (y-a)**2
```

```python
def init_param(dim):
    w=np.random.uniform(0,1,[dim,1])#Vector of 0
    b=0.1
    assert(w.shape == (dim, 1))
    assert(isinstance(b, float) or isinstance(b, int))
    return w,b
```

```python
def propagate(w,b,x,y):
    #forward propagation
    m=x.shape[1]
    z=np.dot(x,w)+b
    a=relu(z)
    cost=np.sum(mse(y,a))/m
    #backward propagation
    dw = (1/m) * np.dot(x.T,((a-y)))
    db = np.longdouble(sum((((1/m) *sum(a-y)))))
    grads={
        "dw":dw,
        "db":db,
    }
    assert(dw.shape == w.shape)
    assert(db.dtype == float)
    cost = np.squeeze(cost)
    assert(cost.shape == ())
    
    return grads,cost
```

```python
def optimize(w, b, x, y, num_iterations, learning_rate, print_cost = False):
    costs=[]
    for i in range(num_iterations):
        grads, cost = propagate(w, b, x, y)
        dw=grads['dw']
        db=grads['db']
        #gradient descent
        w=w-learning_rate*dw
        b=b-learning_rate*db
        #print cost
        if i % 100 == 0:
            costs.append(cost)
    
    params = {"w": w,
              "b": b}
    
    grads = {"dw": dw,
             "db": db}

    return params,grads,cost
```

```python
def predict(w,b,x):
    m = x.shape[1]
    y_prediction = np.zeros((1,m))
    w = w.reshape(x.shape[1], 1)
    z=np.dot(x,w)+b
    a=relu(z)
   
    
    for i in range(a.shape[1]):
        
        y_prediction = np.where(a > 0, a, 0.0)
    
    
    return y_prediction,a
```

```python
def model(x_train, y_train, x_test, y_test, num_iterations = 2000, learning_rate = 0.5, print_cost = False):
    dim2 = x_train.shape[1]
    w, b = init_param(dim2)
    param,grads,cost=optimize(w,b,x_train,y_train,num_iterations,learning_rate,print_cost=False)
    w=param['w']
    b=param['b']
    
    y_prediction_train=predict(w,b,x_train)
    y_prediction_test=predict(w,b,x_test)
    
    print(w)
    print(b)
    print("train accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_train - y_train)) * 100))
    print("test accuracy: {} %".format(100 - np.mean(np.abs(y_prediction_test - y_test)) * 100)) 
    d = {"costs": costs,
         "Y_prediction_test": y_prediction_test, 
         "Y_prediction_train" : y_prediction_train, 
         "w" : w, 
         "b" : b,
         "learning_rate" : learning_rate,
         "num_iterations": num_iterations}    
    return d
```

### D.Optimisation

```python
from time import time
from sklearn.model_selection import GridSearchCV
param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001]}

grid = GridSearchCV(model_final, param_grid, refit = True, scoring='r2',cv=KFold,verbose=3)
grid.fit(X_train,y_train.values.ravel())
start=time()
print(
    "GridSearchCV took %.2f seconds for %d candidate parameter settings."
    % (time() - start, len(grid.cv_results_["params"]))
)
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results["rank_test_score"] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print(
                "Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results["mean_test_score"][candidate],
                    results["std_test_score"][candidate],
                )
            )
            print("Parameters: {0}".format(results["params"][candidate]))
            print("")
report(grid.cv_results_)
```

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV

param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001]}

grid = GridSearchCV(model_final, param_grid, refit = True, scoring='r2',cv=KFold,verbose=3)
grid.fit(X_train,y_train.values.ravel())
start=time()
print(
    "GridSearchCV took %.2f seconds for %d candidate parameter settings."
    % (time() - start, len(grid.cv_results_["params"]))
)
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results["rank_test_score"] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print(
                "Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results["mean_test_score"][candidate],
                    results["std_test_score"][candidate],
                )
            )
            print("Parameters: {0}".format(results["params"][candidate]))
            print("")
print(report(grid.cv_results_))
print(grid.best_params_)

param_grid = {'C':[0.1,1,10,100], 'gamma':[1,0.1,0.01,0.001]}

grid = RandomizedSearchCV(model_final, param_grid, refit = True, scoring='r2',cv=KFold,verbose=3)
grid.fit(X_train,y_train.values.ravel())
start=time()
print(
    "GridSearchCV took %.2f seconds for %d candidate parameter settings."
    % (time() - start, len(grid.cv_results_["params"]))
)
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results["rank_test_score"] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print(
                "Mean validation score: {0:.3f} (std: {1:.3f})".format(
                    results["mean_test_score"][candidate],
                    results["std_test_score"][candidate],
                )
            )
            print("Parameters: {0}".format(results["params"][candidate]))
            print("")
print(report(grid.cv_results_))
print(grid.best_params_)


```
